```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, resample
import scipy.stats as stats
import os

from ITMO_FS.wrappers import RecursiveElimination
from sklearn.svm import SVC

from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate
from sklearn.feature_selection import mutual_info_classif
from sklearn.metrics import accuracy_score, confusion_matrix

feature_counts = {}
model = SVC(kernel='linear')
use_cv = True
echo = False
rf_acc = []
knn_acc = []
MLP_acc = []

rf_std_dev = []
knn_std_dev = []
MLP_std_dev = []

num_features_to_retain = 120

t17.X_train, t17.X_test,  t17.y_train,  t17.y_test = train_test_split(t17.data_array, t17.class_labels, test_size=0.2, random_state=42)

rfe = RecursiveElimination(model, num_features_to_retain)
rfe.fit(np.vstack((t17.X_train, t17.X_test)), np.concatenate((t17.y_train, t17.y_test)))

rf_results  = t17.RForest_train(echo=echo, validation='kfold' if use_cv else 'standard', cv=5)
knn_results =  t17.KNN_train(echo=echo, validation='kfold' if use_cv else 'standard', cv=5)
mlp_results  = t17.MLP_train(hidden_layer_sizes=(200, 100, 50), max_iter=10000, alpha=0.001, validation='kfold' if use_cv else 'standard', cv=5)
```

```python

rf_acc.append(rf_results['mean_accuracy'] if use_cv else rf_results)
knn_acc.append(knn_results['mean_accuracy'] if use_cv else knn_results)
MLP_acc.append(mlp_results['mean_accuracy'] if use_cv else mlp_results)

if use_cv:
    rf_std_dev.append(rf_results['std_accuracy'])
    knn_std_dev.append(knn_results['std_accuracy'])
    MLP_std_dev.append(mlp_results['std_accuracy'])

for feature in rfe.__features__:
    feature_counts[feature] = feature_counts.get(feature, 0) + 1
```

```python
print(knn_std_dev)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
sns.set_theme(style="darkgrid")

sns.lineplot(x=[1,2,3], y=knn_acc, ax=ax1, label='KNN Accuracy')
sns.lineplot(x=[1,2,3], y=rf_acc, ax=ax1, label='Random Forest Accuracy')
sns.lineplot(x=[1,2,3], y=MLP_acc, ax=ax1, label='MLP Accuracy')
ax1.set_title('Accuracy vs Number of Features')
ax1.set_xlabel('Number of features')
ax1.set_ylabel('Accuracy')

if use_cv:
    sns.lineplot(x=[1,2,3], y=knn_std_dev, ax=ax2, label='KNN Std Dev')
    sns.lineplot(x=[1,2,3], y=rf_std_dev, ax=ax2, label='Random Forest Std Dev')
    sns.lineplot(x=[1,2,3], y=MLP_std_dev, ax=ax2, label='MLP Std Dev')
    ax2.set_title('Standard Deviation vs Number of Features')
    ax2.set_xlabel('Number of features')
    ax2.set_ylabel('Standard Deviation')

plt.tight_layout()
plt.show()
```
