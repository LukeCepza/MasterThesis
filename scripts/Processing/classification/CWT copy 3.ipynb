#### Test using all intensities

```python
import copy

from Terminator import Terminator
t16 = Terminator("D:\shared_git\MaestriaThesis\FeaturesTabs\pp04_t16.csv",  method = 'dwt_energy', 
    Types = [1,2,3,4,5,6], mod = ['AirI','AirD','VibI','VibD','CarI','CarD'])
```

```python
t16.check_class_distribution()
```

```python
t16.KNN_train(validation='kfold', cv = 5)
t16.RForest_train(validation='kfold', cv = 5)
t16.MLP_train(hidden_layer_sizes=(200, 100, 50), max_iter=10000, alpha=0.001, validation='kfold', cv=5)
```

```python
t16.KNN_train(validation='kfold', cv = 5)
t16.RForest_train(validation='kfold', cv = 5)
t16.MLP_train(hot_encode = True, hidden_layer_sizes=(200, 100, 50), max_iter=10000, alpha=0.001, validation='kfold', cv=5)
```

```python
t16_entropy = copy.deepcopy(t16)
t16_entropy.feature_select_entropy(plot = True, number_of_f=0, echo = False, threshold=0.0005) #plot all features
```

```python
t16_entropy.KNN_train(validation='kfold', cv = 5)
t16_entropy.RForest_train(validation='kfold', cv = 5)
t16_entropy.MLP_train(hidden_layer_sizes=(200, 100, 50), max_iter=10000, alpha=0.001, validation='kfold', cv=5)
```
